# Sentiment Analysis

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sentanalyzer=SentimentIntensityAnalyzer()
```

```python
data['Scores']=data['Review_Text'].apply(lambda review: sentanalyzer.polarity_scores(review))
```

```python
data
```

```python
data['compound']=data['Scores'].apply(lambda f:f['compound'])
```

```python
data
```

```python
data['compound_posorneg']= data['compound'].apply(lambda score:'Positive' if score >= 0 else 'Negative')
```

```python
data
```

# Text Classification

```python
data=data[['Review_Text','Rating','Rating_posorneg']]
```

```python
data
```

```python
x=data['Review_Text']
```

```python
y=data['Rating_posorneg']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.svm import LinearSVC
```

```python
textclf=Pipeline([("tfidf",TfidfVectorizer()),("clf",LinearSVC())])
```

```python
textclf.fit(x_train,y_train)
```

```python
predictions = textclf.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,predictions))
```

```python
cm=confusion_matrix(y_test,predictions)
cm
```
